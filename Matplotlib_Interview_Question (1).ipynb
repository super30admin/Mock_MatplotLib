```python
import pandas as pd
data=pd.read_csv('data.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep','Oct', 'Nov', 'Dec'])
# y=data[data['Month']=='Dec'].sum()
# print(y)
```

```python
import numpy as np

y1=np.array([9400,10350,11150,11800,11850,12850,13600,14100,13400,13450,12750,12750])
y2=np.array([1410000,10350,1672500,1770000,1777500,1927500,2040000,2115000,2010000,2017500,1912500,1912500])
fig, ax1 = plt.subplots()
ax1.plot(x, y1, marker='s', color='b', label='x1')
ax1.set_xlabel('Months')
ax1.set_ylabel('total travellers', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.legend()
ax2 = ax1.twinx()
ax2.plot(x, y2, marker='x', color='r', label='x2')
ax2.set_ylabel('total revenue', color='r')
ax2.tick_params(axis='y', labelcolor='r')
plt.title('Comparison of total travellers and total revenue')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y1, y2, color='blue', marker='o')
plt.title('Relationship between Travelers and Revenue Generated')
plt.xlabel('Number of Travelers')
plt.ylabel('Revenue Generated')
```
