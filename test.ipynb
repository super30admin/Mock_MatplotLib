```python
import pandas as pd

data = pd.read_csv("travel.csv")
```

```python
print(data)
```

```python
d1 = data.groupby(['Month'])['Travelers'].sum().reset_index()
```

```python
print(d1)
```

```python
d2 = data.groupby(['Month'])['Revenue'].sum().reset_index()
```

```python
d2
```

```python
x = d1['Month'].to_list() 
y = d1['Travelers'].to_list()
```

```python
x1 = d2['Month'].to_list() 
y1 = d2['Revenue'].to_list()
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.plot(x1,y1, '-.')
plt.show()
```

```python
d3 = data.groupby(['Destination'])['Travelers'].sum().reset_index()
```

```python
d3
```

```python
x3 = d3['Destination'].to_list()
y3 = d3['Travelers'].to_list()
```

```python
fig = plt.figure(figsize=(10,7))
plt.pie(y3, labels = x3)
plt.show()
```
